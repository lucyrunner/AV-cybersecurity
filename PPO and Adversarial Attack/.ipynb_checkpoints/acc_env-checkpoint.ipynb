```python
from typing import Optional
import numpy as np
import gymnasium as gym
from gymnasium import spaces

DT = 0.1
TH = 1.5
D0 = 5.0
V_REF = 15.0
A_MIN = -3.5
A_MAX =  2.0
OBS_LOW  = np.array([0.0,  -20.0,  0.0], dtype=np.float32)
OBS_HIGH = np.array([200.0,  20.0, 30.0], dtype=np.float32)
DEFAULT_SEED = 0

RHO = 1.225
C_D = 0.32
A_F = 2.2
C_R = 0.015
M   = 1500.0
G   = 9.81
```

```python
class ACCEnv(gym.Env):
    metadata = {"render_modes": ["human"], "render_fps": int(1 / DT)}

    def __init__(self, normalize_obs: bool = False, seed: Optional[int] = None):
        super().__init__()
        self.normalize_obs = normalize_obs
        self.seed_val = seed if seed is not None else DEFAULT_SEED
        self.np_random, _ = gym.utils.seeding.np_random(self.seed_val)

        self.observation_space = spaces.Box(low=OBS_LOW, high=OBS_HIGH, dtype=np.float32)
        self.action_space = spaces.Box(
            low=np.array([A_MIN], dtype=np.float32),
            high=np.array([A_MAX], dtype=np.float32),
            dtype=np.float32
        )

        self.x_ego, self.x_lead = 0.0, 30.0
        self.v_ego, self.v_lead = V_REF - 0.5, V_REF
        self.a_ego = 0.0

        self.road_grade = 0.0
        self.lead_acc = None
        self.a_lead_est = 0.0
        self._safety_obs_override = None

        self.current_step, self.max_steps, self.collision = 0, 400, False
        self.reset()

    def _resistive_accel(self, v: float, grade: float = 0.0) -> float:
        a_drag  = 0.5 * RHO * C_D * A_F * (v ** 2) / M
        a_roll  = C_R * G * np.cos(np.arctan(grade))
        a_grade = G * np.sin(np.arctan(grade))
        return a_drag + a_roll + a_grade
```

```python
    def _obs_to_array(self):
        return np.array([
            self.x_lead - self.x_ego,
            self.v_lead - self.v_ego,
            self.v_ego
        ], dtype=np.float32)

    def _normalize(self, obs: np.ndarray) -> np.ndarray:
        return (2.0 * (obs - OBS_LOW) / (OBS_HIGH - OBS_LOW + 1e-8) - 1.0).astype(np.float32)

    def _denormalize(self, obs_norm: np.ndarray) -> np.ndarray:
        return (((obs_norm + 1.0) / 2.0) * (OBS_HIGH - OBS_LOW) + OBS_LOW).astype(np.float32)

    def set_safety_obs_for_filter(self, obs):
        self._safety_obs_override = np.array(obs, dtype=np.float32)

    def clear_safety_obs_for_filter(self):
        self._safety_obs_override = None

    def _compute_amax_safe(self, obs_for_filter: np.ndarray) -> float:
        dx, dv, v = float(obs_for_filter[0]), float(obs_for_filter[1]), float(obs_for_filter[2])
        a_lead = float(self.a_lead_est)
        num = dx - TH * v + (dv - TH * a_lead) * DT
        denom = TH * DT
        if denom <= 0:
            return A_MIN
        a_max_safe = num / (denom + 1e-8)
        return float(np.clip(a_max_safe, A_MIN, A_MAX))

    def _apply_safety(self, a_rl: float, dx: float, dv: float, v: float) -> float:
        a_safe_max = self._compute_amax_safe(np.array([dx, dv, v], dtype=np.float32))
        a_clamped = min(a_rl, a_safe_max)
        return float(np.clip(a_clamped, A_MIN, A_MAX))
```

```python
    def step(self, action):
        a = float(np.array(action).reshape(-1)[0]) if isinstance(action, (list, tuple, np.ndarray)) else float(action)

        if self._safety_obs_override is not None:
            obs_for_filter = self._safety_obs_override
            if np.max(np.abs(obs_for_filter)) <= 1.1 and self.normalize_obs:
                obs_for_filter = self._denormalize(obs_for_filter)
            self._safety_obs_override = None
        else:
            obs_for_filter = self._obs_to_array()

        a_clamped = self._apply_safety(a, obs_for_filter[0], obs_for_filter[1], obs_for_filter[2])
        a_resist = self._resistive_accel(self.v_ego, self.road_grade)
        a_net = a_clamped - a_resist

        self.x_ego += self.v_ego * DT + 0.5 * a_net * DT * DT
        self.v_ego = float(np.clip(self.v_ego + a_net * DT, 0.0, 100.0))
        self.a_ego = a_clamped

        if self.lead_acc is not None:
            self.v_lead = float(np.clip(self.v_lead + self.lead_acc * DT, 0.0, 100.0))
            self.x_lead += self.v_lead * DT + 0.5 * self.lead_acc * DT * DT
        else:
            self.x_lead += self.v_lead * DT

        self.current_step += 1

        # ---- ADD PRINTS HERE ----
        if self.current_step % 20 == 0:  # print every 20 steps to reduce spam
            print(f"[step {self.current_step}] a_rl={a:.3f}, a_clamped={a_clamped:.3f}, "
                  f"a_resist={a_resist:.3f}, a_net={a_net:.3f}, v_ego={self.v_ego:.3f}, "
                  f"dx={self.x_lead - self.x_ego:.3f}")

        obs = self._obs_to_array()
        collision = obs[0] <= 0.0
        if collision:
            self.collision = True

        info = {
            "collision": collision,
            "dx": float(obs[0]),
            "ego_v": float(obs[2]),
            "lead_v": float(self.v_lead),
            "applied_action": float(a_clamped),
            "resist_accel": float(a_resist),
            "net_accel": float(a_net),
        }

        speed_err = (self.v_ego - V_REF)
        dsafe = D0 + TH * self.v_ego
        safe_pen = max(0.0, dsafe - obs[0])
        action_pen = 0.5 * (a_clamped ** 2)
        reward = -(0.5 * (speed_err ** 2) + 2.0 * (safe_pen ** 2) + 0.01 * action_pen)

        done = self.collision
        truncated = (self.current_step >= self.max_steps)
        if self.collision:
            info["terminal_reason"] = "collision"
        elif truncated:
            info["terminal_reason"] = "time_limit"

        obs_out = self._normalize(obs) if self.normalize_obs else obs
        return obs_out, float(reward), bool(done), bool(truncated), info
```

```python
    def reset(self, *, seed: Optional[int] = None, options: Optional[dict] = None):
        if seed is not None:
            self.seed_val = int(seed)
            self.np_random, _ = gym.utils.seeding.np_random(self.seed_val)

        self.current_step, self.collision = 0, False
        self.x_ego, self.x_lead = 0.0, 30.0 + self.np_random.uniform(-1.0, 1.0)
        self.v_ego = V_REF - 0.5 + self.np_random.uniform(-0.2, 0.2)
        self.v_lead = V_REF + self.np_random.uniform(-0.5, 0.5)
        self.a_ego, self._safety_obs_override = 0.0, None

        obs = self._obs_to_array()
        # --- THIS LINE IS ESSENTIAL ---
        return (self._normalize(obs) if self.normalize_obs else obs), {}
```

```python
import numpy as np
import gymnasium as gym
from gymnasium import spaces

class ACCEnv(gym.Env):
    metadata = {"render_modes": []}

    def __init__(self,
                 dt=0.05,
                 d0=5.0,          # standstill spacing [m]
                 h=1.2,           # time headway [s]
                 v_min=0.0,
                 v_max=40.0,
                 a_min=-5.0,
                 a_max=3.0,
                 z_clip=200.0):
        super().__init__()
        # Parameters
        self.dt = float(dt)
        assert self.dt > 0, "dt must be > 0"
        self.d0 = float(d0)
        self.h = float(h)
        self.v_min = float(v_min)
        self.v_max = float(v_max)
        self.min_acc = float(a_min)
        self.max_acc = float(a_max)
        self.z_clip = float(z_clip)

        # State: [z, v_rel, v_e]  (spacing error, relative vel (v_l - v_e), ego vel)
        high = np.array([z_clip, 50.0, v_max], dtype=np.float32)
        low  = np.array([-z_clip, -50.0, v_min], dtype=np.float32)
        self.observation_space = spaces.Box(low=low, high=high, dtype=np.float32)
        self.action_space = spaces.Box(low=np.array([self.min_acc], np.float32),
                                       high=np.array([self.max_acc], np.float32),
                                       dtype=np.float32)

        self.state = None
        self.t = 0.0
        self.ep_len = 0
        self.max_steps = int(60.0 / self.dt)  # 60s horizon

    # --- Lead vehicle profile (edit as needed) ---
    def lead_acc(self, t: float) -> float:
        # Smooth bounded sinusoid
        return 0.5 * np.sin(0.2 * t)

    def lead_vel(self, v_l_prev: float, t: float) -> float:
        v_l = v_l_prev + self.lead_acc(t) * self.dt
        return float(np.clip(v_l, self.v_min, self.v_max))

    # --- Gymnasium API ---
    def reset(self, *, seed=None, options=None):
        super().reset(seed=seed)
        self.t = 0.0
        self.ep_len = 0
        v_e0 = 15.0
        v_l0 = 15.5
        d0 = self.d0 + self.h * v_e0  # desired distance
        # spacing error z = d_actual - (d0 + h * v_e), initialize near 0
        z0 = 0.0
        v_rel0 = v_l0 - v_e0
        self.state = np.array([z0, v_rel0, v_e0], dtype=np.float32)
        self._v_l = float(v_l0)
        obs = self.state.copy()
        return obs, {}

    def step(self, action):
        # Unpack & clip action
        if np.isscalar(action):
            u = float(action)
        else:
            u = float(np.array(action, dtype=np.float32).squeeze())
        u = float(np.clip(u, self.min_acc, self.max_acc))

        z, v_rel, v_e = map(float, self.state)
        # Update lead car
        self._v_l = self.lead_vel(self._v_l, self.t)
        a_l = self.lead_acc(self.t)

        # Continuous dynamics (Euler)
        # z = d - (d0 + h * v_e), so z_dot = v_rel - h * u
        z_dot = v_rel - self.h * u
        v_rel_dot = a_l - u
        v_e_dot = u

        z  = z + self.dt * z_dot
        v_rel = v_rel + self.dt * v_rel_dot
        v_e = np.clip(v_e + self.dt * v_e_dot, self.v_min, self.v_max)

        self.state = np.array([z, v_rel, v_e], dtype=np.float32)
        self.state = np.nan_to_num(self.state, nan=0.0, posinf=1e6, neginf=-1e6)

        # Reward (tune weights)
        wz, wa, wj = 1.0, 0.01, 0.001
        # Estimate jerk via finite difference of action (store last action)
        if not hasattr(self, "_u_prev"):
            self._u_prev = u
        jerk = (u - self._u_prev) / self.dt
        self._u_prev = u

        reward = - (wz * (z**2) + wa * (u**2) + wj * (jerk**2))

        # Termination / truncation
        self.t += self.dt
        self.ep_len += 1

        # Collision if actual distance d = z + (d0 + h*v_e) ≤ 0
        d_actual = z + (self.d0 + self.h * v_e)
        collision = d_actual <= 0.0

        terminated = bool(collision)
        truncated = bool(self.ep_len >= self.max_steps)

        info = {
            "d_actual": float(d_actual),
            "v_l": float(self._v_l),
            "a_l": float(a_l),
            "u": float(u),
        }

        return self.state.copy(), float(reward), terminated, truncated, info
```

```python
if __name__ == "__main__":
    env = ACCEnv()
    obs, info = env.reset()
    print("Initial obs:", obs)

    total_reward = 0.0
    for step in range(300):  # run 300 simulation steps
        action = env.action_space.sample()
        obs, reward, terminated, truncated, info = env.step(action)
        total_reward += reward

        if step % 50 == 0:
            print(f"step={step:03d}  z={obs[0]:6.2f}  v_rel={obs[1]:6.2f}  v_e={obs[2]:6.2f}  a={info['u']:5.2f}")

        if terminated or truncated:
            print("Episode ended early.")
            break

    print(f"Final obs: {obs}")
    print(f"Total reward: {total_reward:.2f}")
```
